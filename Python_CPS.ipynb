```python
from google.colab import userdata
import requests
import pandas as pd # tabluar data
import plotly.express as px

CENSUS_API_KEY = userdata.get('CENSUS_API_KEY')
```

# Functions

Will convert the API response to a Pandas Dataframe

```python
def json_to_dataframe(response):
    return pd.DataFrame(response.json()[1:], columns=response.json()[0])
```

# Example

```python
url = "https://api.census.gov/data/2022/cps/foodsec/dec?get=HES1A,HRFS12M1,HRFS30D5&key=" +CENSUS_API_KEY
```

```python
response = requests.request("GET", url)
```

```python
json_to_dataframe(requests.request("GET", url))
```
